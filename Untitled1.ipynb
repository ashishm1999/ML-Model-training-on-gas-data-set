```python
print("TIME-SERIES DECOMPOSITION")


from statsmodels.tsa.seasonal import seasonal_decompose

# Downsample for decomposition (computational efficiency)
downsample_factor = max(10, len(df) // 5000)
df_downsampled = df.iloc[::downsample_factor].copy()
print(f"Downsampled from {len(df):,} to {len(df_downsampled):,} points")

# Determine appropriate period
period = min(50, len(df_downsampled) // 3)
print(f"Using period: {period}")

try:
    decomposition = seasonal_decompose(
        df_downsampled['Sensor_Reading'],
        model='additive',  # Use 'multiplicative' if variance increases with level
        period=period
    )

    fig, axes = plt.subplots(4, 1, figsize=(15, 10))

    decomposition.observed.plot(ax=axes[0], color='steelblue')
    axes[0].set_title('Original Time Series', fontsize=14, fontweight='bold')
    axes[0].set_ylabel('Observed')

    decomposition.trend.plot(ax=axes[1], color='orange')
    axes[1].set_title('Trend Component', fontsize=14, fontweight='bold')
    axes[1].set_ylabel('Trend')

    decomposition.seasonal.plot(ax=axes[2], color='green')
    axes[2].set_title('Seasonal Component', fontsize=14, fontweight='bold')
    axes[2].set_ylabel('Seasonal')

    decomposition.resid.plot(ax=axes[3], color='red')
    axes[3].set_title('Residual Component', fontsize=14, fontweight='bold')
    axes[3].set_ylabel('Residual')
    axes[3].set_xlabel('Time Step')

    for ax in axes:
        ax.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig(f"{RESULTS_DIR}/02_decomposition.png", dpi=150, bbox_inches='tight')
    plt.show()

    print(" Decomposition successful!")

except Exception as e:
    print(f"Decomposition failed: {e}")
    print("Continuing with analysis...")
```

```python
print("MOVING AVERAGE MODELS")


from sklearn.metrics import mean_absolute_error, mean_squared_error

# Use subset for modeling
subset_size = min(10000, len(df))
data = df['Sensor_Reading'].iloc[:subset_size].values

# Split train/test (80-20)
train_size = int(subset_size * 0.8)
train_data = data[:train_size]
test_data = data[train_size:]

print(f"Train: {len(train_data):,} | Test: {len(test_data):,}")

# Test multiple window sizes
window_sizes = [5, 10, 20]
results = []

for window in window_sizes:
    print(f"  Processing MA-{window}...")

    # Generate forecasts for test set
    forecasts = []
    for i in range(len(test_data)):
        if i == 0:
            # First prediction: use last 'window' points from training
            recent = train_data[-window:]
        else:
            # Subsequent predictions: use rolling window including test points
            if i >= window:
                recent = test_data[i-window:i]
            else:
                # Combine training and test data
                recent = np.concatenate([
                    train_data[-(window-i):],
                    test_data[:i]
                ])

        forecasts.append(np.mean(recent))

    # Calculate metrics
    mae = mean_absolute_error(test_data, forecasts)
    rmse = np.sqrt(mean_squared_error(test_data, forecasts))

    results.append({
        'Model': f'MA-{window}',
        'Window': window,
        'MAE': mae,
        'RMSE': rmse
    })

# Display results
results_df = pd.DataFrame(results)
print("\n" + "="*60)
print("MOVING AVERAGE RESULTS")
print("="*60)
print(results_df.to_string(index=False))
print("="*60)

```

```python
print("LINEAR REGRESSION")


def create_features(data, n_lags=20):
    """
    Create supervised learning features from time-series.

    For each time point t, create features:
    - Lags: data[t-1], data[t-2], ..., data[t-n_lags]
    - Rolling mean (5-point window)
    - Rolling std (5-point window)
    - Rolling mean (10-point window)
    - Rolling std (10-point window)
    - Time index

    Target: data[t]
    """
    features = []
    targets = []

    for i in range(n_lags, len(data)):
        # Extract lag features
        lags = data[i-n_lags:i]

        # Calculate rolling statistics
        rolling_mean_5 = np.mean(data[i-5:i])
        rolling_std_5 = np.std(data[i-5:i])
        rolling_mean_10 = np.mean(data[i-10:i])
        rolling_std_10 = np.std(data[i-10:i])

        # Combine all features
        feature_vec = list(lags) + [
            rolling_mean_5, rolling_std_5,
            rolling_mean_10, rolling_std_10,
            i  # Time index
        ]

        features.append(feature_vec)
        targets.append(data[i])

    return np.array(features), np.array(targets)

# Create features
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

X, y = create_features(data, n_lags=20)

print(f"Feature matrix shape: {X.shape}")
print(f"Number of features: {X.shape[1]}")
print(f"  - Lag features: 20")
print(f"  - Rolling statistics: 4")
print(f"  - Time index: 1")

# Split data
train_size_lr = int(len(X) * 0.8)
X_train = X[:train_size_lr]
X_test = X[train_size_lr:]
y_train = y[:train_size_lr]
y_test = y[train_size_lr:]

print(f"\nTraining samples: {len(X_train):,}")
print(f"Test samples: {len(X_test):,}")

# Feature scaling (important for regularization)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\n Features standardized (mean=0, std=1)")

# Train model
model = LinearRegression()
model.fit(X_train_scaled, y_train)

print(" Model trained")

# Make predictions
y_pred = model.predict(X_test_scaled)

# Evaluate
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = model.score(X_test_scaled, y_test)

print("\n" + "="*60)
print("LINEAR REGRESSION RESULTS")
print("="*60)
print(f"R² Score:  {r2:.4f}")
print(f"MAE:       {mae:.4f}")
print(f"RMSE:      {rmse:.4f}")
print("="*60)
```

```python
print("LSTM MODEL")


import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping

print(f"TensorFlow version: {tf.__version__}")
print(f"GPUs available: {len(tf.config.list_physical_devices('GPU'))}")

# Set seed for reproducibility
tf.random.set_seed(SEED)

# Prepare sequences
def create_sequences(data, seq_len=50):
    """
    Transform time-series into sequences.

    Input: [v1, v2, v3, v4, v5, v6, ...]
    Output:
        X: [[v1, v2, ..., v50], [v2, v3, ..., v51], ...]
        y: [v51, v52, ...]
    """
    X, y = [], []
    for i in range(seq_len, len(data)):
        X.append(data[i-seq_len:i])
        y.append(data[i])
    return np.array(X), np.array(y)

# Normalize data (LSTM works better with scaled inputs)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
data_normalized = scaler.fit_transform(data.reshape(-1, 1)).flatten()

print(" Data normalized to [0, 1] range")

# Create sequences
seq_length = 50  # Use last 50 time steps to predict next value
X, y = create_sequences(data_normalized, seq_length)

# Reshape for LSTM: (samples, timesteps, features)
X = X.reshape((X.shape[0], X.shape[1], 1))

print(f"\nSequence shape: {X.shape}")
print(f"  - Samples: {X.shape[0]:,}")
print(f"  - Timesteps: {X.shape[1]}")
print(f"  - Features: {X.shape[2]}")

# Split data
train_size_lstm = int(len(X) * 0.8)
X_train_lstm = X[:train_size_lstm]
X_test_lstm = X[train_size_lstm:]
y_train_lstm = y[:train_size_lstm]
y_test_lstm = y[train_size_lstm:]

print(f"\nTraining sequences: {len(X_train_lstm):,}")
print(f"Test sequences: {len(X_test_lstm):,}")
```

```python
# Define architecture
model_lstm = keras.Sequential([
    # First LSTM layer (returns sequences for stacking)
    layers.LSTM(64, activation='relu', return_sequences=True,
                input_shape=(seq_length, 1)),
    layers.Dropout(0.2),  # Prevent overfitting

    # Second LSTM layer
    layers.LSTM(32, activation='relu'),
    layers.Dropout(0.2),

    # Dense layers
    layers.Dense(16, activation='relu'),
    layers.Dense(1)  # Output layer (single value prediction)
])

# Compile model
model_lstm.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='mse',      # Mean Squared Error
    metrics=['mae']  # Monitor MAE during training
)

print("\nModel Architecture:")
model_lstm.summary()
```

```python
print("\nTraining LSTM...")
print("This may take 5-10 minutes depending on hardware...")

# Early stopping to prevent overfitting
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True,
    verbose=1
)

# Train
history = model_lstm.fit(
    X_train_lstm, y_train_lstm,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

print("\n Training complete!")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Loss curve
axes[0].plot(history.history['loss'], label='Training Loss', linewidth=2)
axes[0].plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
axes[0].set_title('Model Loss During Training', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss (MSE)')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# MAE curve
axes[1].plot(history.history['mae'], label='Training MAE', linewidth=2)
axes[1].plot(history.history['val_mae'], label='Validation MAE', linewidth=2)
axes[1].set_title('Model MAE During Training', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('MAE')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(f"{RESULTS_DIR}/05_lstm_history.png", dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Predict on test set
y_pred_lstm = model_lstm.predict(X_test_lstm, verbose=0)

# Inverse transform to original scale
y_test_original = scaler.inverse_transform(y_test_lstm.reshape(-1, 1)).flatten()
y_pred_original = scaler.inverse_transform(y_pred_lstm).flatten()

# Calculate metrics
mae_lstm = mean_absolute_error(y_test_original, y_pred_original)
rmse_lstm = np.sqrt(mean_squared_error(y_test_original, y_pred_original))

print("="*60)
print("LSTM TEST PERFORMANCE")
print("="*60)
print(f"MAE:  {mae_lstm:.4f}")
print(f"RMSE: {rmse_lstm:.4f}")
print("="*60)
```
